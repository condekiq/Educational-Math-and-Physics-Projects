```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-2.0, 1.0, 1000)
Y = np.linspace(-1.0, 1.0, 1000)

X_grid, Y_grid = np.meshgrid(X, Y)
Z_grid = X_grid + Y_grid*1j

def sequence(c_grid_input, num):
    c_grid_input_shape = np.shape(c_grid_input)
    c_grid_output = 0*c_grid_input

    for i in np.arange(c_grid_input_shape[0]):
        for j in np.arange(c_grid_input_shape[1]):
                z = 0.0 + 0.0*1j
                num_aux = num
                while (np.abs(z) <= 2 and num_aux != 0):
                    z = z**2 + c_grid_input[i][j]
                    num_aux = num_aux - 1
                c_grid_output[i][j] = z
    return c_grid_output
```

```python
Z_hit = np.abs(sequence(Z_grid, 30))

for j in np.arange(len(X)):
    for i in np.arange(len(Y)):
        if Z_hit[i][j] > 2:
            Z_hit[i][j] = 1
        else:
            Z_hit[i][j] = -1
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.pcolormesh(X_grid, Y_grid, Z_hit)
ax.grid(True)
```
